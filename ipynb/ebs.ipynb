```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver = webdriver.Edge()
driver.get('https://www.ebsi.co.kr/ebs/pot/potn/retrieveTchrSubMain.ebs')
```

```python

list_ebs = []
tea_ind = 10
no = 1
year = 23
toggle_num = 2
import time
while 1:
    print("크롤링 시작합니다.")
    try:
        driver.find_element(by='xpath',value =f'/html/body/div[2]/section/div/div[1]/div[2]/ul/li[3]/a').click()
        print("시작1")
        ind = 6
        driver.find_element(by='xpath',value =f'/html/body/div[2]/section/div/div[2]/div[2]/div/div/div[{tea_ind}]/div/a/div[2]/span' ).is_displayed()
        print("선생님을 찾았습니다.")
        # 선생님 이름
        teacher = driver.find_element(by='xpath',value =f'/html/body/div[2]/section/div/div[2]/div[2]/div/div/div[{tea_ind}]/div/a/div[2]/span' ).text
        time.sleep(1)
        driver.find_element(by='xpath',value =f'/html/body/div[2]/section/div/div[2]/div[2]/div/div/div[{tea_ind}]/div/a/div[2]/span').click()
        print(teacher)
        time.sleep(0.5)
        # 하트수
        heart = driver.find_element(by='xpath',value =f'//*[@id="spanCntGoodYn"]' ).text
        # 수강후기 클릭
        driver.find_element(by='xpath',value =f'//*[@id="cdTabMnuArea"]/li[6]/a' ).click()
        time.sleep(0.5)
        print("강의 리스트를 첫번째로 변경합니다.")
        # 강좌 토글버튼 선택
        toggle_num = 2
        driver.find_element(by='xpath',value =f'//*[@id="gotoTab"]/div/div[1]/div/select[1]/option[{toggle_num}]' ).click()
        time.sleep(3)
        now_page = 1
        tea_ind +=1
        while 1:
            print("두번째 while문으로 들어갑니다.")
            if int(year) == 23:
                try:
                    print("크롤링을 진행합니다.")
                    driver.find_element(by='xpath',value =f'//*[@id="gotoTab"]/div/div[3]/ul/li[{ind}]/div[4]/div/a' ).is_displayed()
                    driver.find_element(by='xpath',value =f'//*[@id="gotoTab"]/div/div[3]/ul/li[{ind}]/div[4]/div/a' ).click()
                    time.sleep(1)
                    reviewer = driver.find_element(by='xpath',value ='//*[@id="gotoTab"]/div/div[2]/div/div[1]/div[1]/ul/li[1]/span' ).text
                    date = driver.find_element(by='xpath',value ='//*[@id="gotoTab"]/div/div[2]/div/div[1]/div[1]/ul/li[2]/span' ).text
                    kinds = driver.find_element(by='xpath',value ='//*[@id="gotoTab"]/div/div[2]/div/div[1]/div[1]/ul/li[3]/span' ).text
                    # 강의명
                    content_title = driver.find_element(by='xpath',value ='//*[@id="gotoTab"]/div/div[2]/div/div[1]/div/p[1]/a' ).text
                    # 리뷰제목
                    review_title = driver.find_element(by='xpath',value ='//*[@id="gotoTab"]/div/div[2]/div/div[1]/div/p[2]' ).text
                    content = driver.find_element(by='xpath',value ='//*[@id="gotoTab"]/div/div[2]/div/div[2]' ).text
                    rated = driver.find_element(by='xpath',value ='//*[@id="gotoTab"]/div/div[2]/div/div[1]/div/div/span' ).text
                    year = date.split('.')[0]
                    if int(year) == 23:
                        print(f"no{no} : 리스트에 내용을 추가합니다.")
                        list_ebs.append([teacher, heart, reviewer, date, kinds,content_title,review_title,  content, rated])
                    time.sleep(1)
                    # 목록으로 돌아감
                    driver.find_element(by='xpath',value ='//*[@id="gotoTab"]/div/div[3]/button' ).click()
                    time.sleep(1)
                    ind += 1
                    print(no)
                    no += 1
                except:
                    print("크롤링이 진행되지 않아 다음 페이지로 전환합니다.")
                    time.sleep(1)
                    total_page = driver.find_element(by='xpath',value =f'//*[@id="gotoTab"]/div/div[4]/div/div/a[2]' ).text
                    if now_page < int(total_page):
                        ind = 2
                        driver.find_element(by='xpath',value =f'//*[@id="gotoTab"]/div/div[4]/div/a[3]' ).click()
                        time.sleep(1)
                        now_page += 1
                        continue
                    elif driver.find_element(by='xpath',value =f'//*[@id="gotoTab"]/div/div[1]/div/select[1]/option[{toggle_num}]' ).is_displayed() == True:
                        toggle_num += 1
                        print(f"다음 {toggle_num}번째 토글로 이동합니다.")
                        driver.find_element(by='xpath',value =f'//*[@id="gotoTab"]/div/div[1]/div/select[1]/option[{toggle_num}]' ).click()
                        time.sleep(3)
                        year = 23
                        continue
                    # 더이상 이동불가한 경우
                    else:
                        print("더이상 페이지를 이동할 수 없어 크롤링을 종료합니다.")
                        ## 선생님 선택 창으로 이동
                        break
            # 작성년도가 23년이 아닌경우
            else:
                try :
                    # 토글을 다음번으로 이동
                    toggle_num += 1
                    print("다음번 강좌로 이동합니다.")
                    driver.find_element(by='xpath',value =f'//*[@id="gotoTab"]/div/div[1]/div/select[1]/option[{toggle_num}]' ).click()
                    time.sleep(3)
                    year = 23
                    continue
                except :
                    # 선생님 선택으로 돌아감
                    driver.find_element(by='xpath',value =f'/html/body/div[2]/section/div/div[1]/div[2]/ul/li[3]/a' ).click()
                    time.sleep(3)
                    print("선생님 선택창으로 돌아갑니다.")
                    year = 23
                    break
            continue
    except:
        print("선생님을 찾을 수 없습니다.")
        break
```
